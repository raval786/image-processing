```python
img = cv2.imread('Data/00-puppy.jpg')
```

# in matplotlib color order is RGB  RED GREEN BLUE
# in opencv color order is BLUE GREEN RED

```python
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # CONVERT BGR TO RGB COLOR CHANNEL
```

```python
plt.imshow(fix_img)
```

```python
img_gray = cv2.imread('DATA/00-puppy.jpg',cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img_gray,cmap='gray')
```

```python
fix_img.shape
```

```python
new_img = cv2.resize(fix_img,(1000,400))
```

```python
plt.imshow(new_img)
```

```python
w_ratio = 0.5 # width ratio  # 50% of orignal
h_ratio = 0.2 # hight ratio  # 20% of orignal
```

```python
new_img = cv2.resize(fix_img,(0,0),fix_img,w_ratio,h_ratio)
```

```python
plt.imshow(new_img)
```

```python
new_img.shape #  ratio
```

```python
new_img = cv2.flip(fix_img,0)
```

```python
plt.imshow(new_img)
```
